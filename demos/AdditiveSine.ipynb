```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

buffer=[]
buffer2=[]

addSinesBuffer=[]

fs=2000

f1=1
f2=3
f3=5
f4=7

Amp1=1
Amp2=0.6
Amp3=0.5
Amp4=0.4

for h in range(0,fs):
    sine1_sample=Amp1*np.sin(2*np.pi/fs*h*f1)
    sine2_sample=Amp2*(np.sin(2*np.pi/fs*h*f2))
    sine3_sample=Amp3*(np.sin(2*np.pi/fs*h*f3))
    sine4_sample=Amp4*(np.sin(2*np.pi/fs*h*f4))
   
    buffer.append(sine1_sample)
    buffer2.append(sine2_sample)
    
    addSines=sine1_sample+sine2_sample+sine3_sample+sine4_sample 
    addSinesBuffer.append(addSines)
    
plt.plot(buffer, color='red')
plt.plot(buffer2,color='blue')
plt.title('Sine Wave')
plt.show()

plt.plot(addSinesBuffer)
plt.title('Square Wave')
plt.show()
```
